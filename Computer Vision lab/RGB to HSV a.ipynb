```python
from skimage import io, color
import matplotlib.pyplot as plt

# Read image
img = io.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')

# Convert to HSV
img_hsv = color.rgb2hsv(img)

# Convert back to RGB
img_rgb = color.hsv2rgb(img_hsv)

# Show both figures side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(img_hsv)
axes[0].set_title("HSV Image")
axes[0].axis("off")

axes[1].imshow(img_rgb)
axes[1].set_title("RGB Image")
axes[1].axis("off")

plt.show()
```
