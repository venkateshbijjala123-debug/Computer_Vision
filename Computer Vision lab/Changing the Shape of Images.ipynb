```python
import cv2
import matplotlib.pyplot as plt

# Read image
image_path = r'C:\Users\venka\Downloads\Computer Vision lab\IMG_1626.jpg'
image = cv2.imread(image_path)

# --- EROSION ---
s1, s2, s3 = 0, 10, 10
t1, t2, t3 = cv2.MORPH_RECT, cv2.MORPH_CROSS, cv2.MORPH_ELLIPSE

tmp1 = cv2.getStructuringElement(t1, (2*s1+1, 2*s1+1), (s1, s1))
tmp2 = cv2.getStructuringElement(t2, (2*s2+1, 2*s2+1), (s2, s2))
tmp3 = cv2.getStructuringElement(t3, (2*s3+1, 2*s3+1), (s3, s3))

erosion1 = cv2.erode(image, tmp1)
erosion2 = cv2.erode(image, tmp2)
erosion3 = cv2.erode(image, tmp3)

# --- DILATION ---
d1, d2, d3 = 0, 10, 20
t1, t2, t3 = cv2.MORPH_RECT, cv2.MORPH_CROSS, cv2.MORPH_ELLIPSE

tmp1 = cv2.getStructuringElement(t1, (2*d1+1, 2*d1+1), (d1, d1))
tmp2 = cv2.getStructuringElement(t2, (2*d2+1, 2*d2+1), (d2, d2))
tmp3 = cv2.getStructuringElement(t3, (2*d3+1, 2*d3+1), (d3, d3))

dilation1 = cv2.dilate(image, tmp1)
dilation2 = cv2.dilate(image, tmp2)
dilation3 = cv2.dilate(image, tmp3)

# --- Display results ---
titles = ["Original", "Erosion Rect", "Erosion Cross", "Erosion Ellipse",
          "Dilation Rect", "Dilation Cross", "Dilation Ellipse"]
images = [image, erosion1, erosion2, erosion3, dilation1, dilation2, dilation3]

plt.figure(figsize=(14, 8))
for i in range(len(images)):
    plt.subplot(2, 4, i+1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
