```python
import cv2
import matplotlib.pyplot as plt

# Read image
image_path = r'C:\Users\venka\Downloads\Computer Vision lab\IMG_1626.jpg'
image_Original = cv2.imread(image_path)

# Apply different blurs
image_MedianBlur = cv2.medianBlur(image_Original, 9)
image_GaussianBlur = cv2.GaussianBlur(image_Original, (9, 9), 10)
image_BilateralBlur = cv2.bilateralFilter(image_Original, 9, 100, 75)

# Show images side by side
titles = ["Original", "Median Blur", "Gaussian Blur", "Bilateral Blur"]
images = [image_Original, image_MedianBlur, image_GaussianBlur, image_BilateralBlur]

plt.figure(figsize=(12, 6))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
