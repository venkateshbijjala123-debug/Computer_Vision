```python
from skimage import io, exposure
import matplotlib.pyplot as plt

# Load image (remembered path)
img = io.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')

# Apply gamma correction
gamma_corrected1 = exposure.adjust_gamma(img, 0.5)  # brighten
gamma_corrected2 = exposure.adjust_gamma(img, 5)    # darken

# Show both results side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(gamma_corrected1)
axes[0].set_title("Gamma 0.5 (Brightened)")
axes[0].axis("off")

axes[1].imshow(gamma_corrected2)
axes[1].set_title("Gamma 5 (Darkened)")
axes[1].axis("off")

plt.show()
```
