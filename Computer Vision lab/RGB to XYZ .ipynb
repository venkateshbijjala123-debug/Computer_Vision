```python
from skimage import io, color
import matplotlib.pyplot as plt
import numpy as np

# Read image
img = io.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')

# Convert to XYZ
img_xyz = color.rgb2xyz(img)

# Convert back to RGB
img_rgb = color.xyz2rgb(img_xyz)

# Clip XYZ values to [0,1] for display
img_xyz_clipped = np.clip(img_xyz, 0, 1)

# Show both figures side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(img_xyz_clipped)
axes[0].set_title("XYZ Image (clipped)")
axes[0].axis("off")

axes[1].imshow(img_rgb)
axes[1].set_title("RGB Image (from XYZ)")
axes[1].axis("off")

plt.show()
```
