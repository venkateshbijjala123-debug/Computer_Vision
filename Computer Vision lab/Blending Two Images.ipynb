```python
import cv2
import matplotlib.pyplot as plt

# Read images
img1 = cv2.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')
img2 = cv2.imread(r'C:\Users\venka\Downloads\Computer Vision lab\IMG_1626.jpg')

# Resize to same size
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Blend images
final_image = cv2.addWeighted(img1, 0.3, img2, 0.7, 0.0)

# Show result
plt.imshow(cv2.cvtColor(final_image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
