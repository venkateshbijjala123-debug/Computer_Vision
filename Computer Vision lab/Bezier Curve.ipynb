```python
from skimage import io, draw
import matplotlib.pyplot as plt

# Load image (remembered path)
img = io.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')

# Get image dimensions
h, w = img.shape[:2]

# Define Bézier curve coordinates safely within bounds
rr, cc = draw.bezier_curve(0, 0, h//2, w//2, h-1, w-1, 100)

# Draw curve (black pixels)
img[rr, cc] = 0

# Show image
plt.imshow(img)
plt.axis("off")
plt.show()
```
