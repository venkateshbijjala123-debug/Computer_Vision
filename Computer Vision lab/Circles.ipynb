```python
from skimage import io, draw
import matplotlib.pyplot as plt

# Load image (remembered path)
img = io.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')

# Define circle center and radius
rr, cc = draw.disk((100, 100), 50, shape=img.shape)  # center=(100,100), radius=50

# Draw circle (black fill)
img[rr, cc] = 0

# Show image
plt.imshow(img)
plt.axis("off")
plt.show()
```
