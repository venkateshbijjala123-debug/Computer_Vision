```python
import cv2
import matplotlib.pyplot as plt

# Read image in grayscale (important for thresholding)
image_path = r'C:\Users\venka\Downloads\Computer Vision lab\IMG_1626.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply different thresholds
_, img1 = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY)
_, img2 = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY_INV)
_, img3 = cv2.threshold(image, 50, 255, cv2.THRESH_TRUNC)
_, img4 = cv2.threshold(image, 50, 255, cv2.THRESH_TOZERO)
_, img5 = cv2.threshold(image, 50, 255, cv2.THRESH_TOZERO_INV)

# Show results side by side
titles = ["Binary", "Binary Inverted", "Truncated", "To Zero", "To Zero Inverted"]
images = [img1, img2, img3, img4, img5]

plt.figure(figsize=(12, 6))
for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
