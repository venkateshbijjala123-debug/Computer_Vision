```python
from skimage import io, color
import numpy as np

# Read image (remembered path)
img = io.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')

# Convert to YPbPr
img_ypbpr = color.rgb2ypbpr(img)

# Convert back to RGB
img_rgb = color.ypbpr2rgb(img_ypbpr)

# Clip YPbPr values into [0,1] and scale to 0–255 for saving
img_ypbpr_clipped = np.clip(img_ypbpr, 0, 1)
io.imsave("image_ypbpr.jpg", (img_ypbpr_clipped * 255).astype(np.uint8))

# Save the reconstructed RGB image safely
img_rgb_clipped = np.clip(img_rgb, 0, 1)
io.imsave("image_rgb_from_ypbpr.jpg", (img_rgb_clipped * 255).astype(np.uint8))
```
