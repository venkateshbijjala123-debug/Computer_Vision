```python
from skimage import io, color
import matplotlib.pyplot as plt
import numpy as np

# Read image
img = io.imread(r"C:\Users\venka\Downloads\Computer Vision lab\images.jpg")

# Convert to YIQ
img_yiq = color.rgb2yiq(img)

# Convert back to RGB
img_rgb = color.yiq2rgb(img_yiq)

# Clip values into [0,1] for safe display
img_yiq_clipped = np.clip(img_yiq, 0, 1)

# Show both figures side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(img_yiq_clipped)
axes[0].set_title("YIQ Image (clipped)")
axes[0].axis("off")

axes[1].imshow(img_rgb)
axes[1].set_title("RGB Image (from YIQ)")
axes[1].axis("off")

plt.show()
```
