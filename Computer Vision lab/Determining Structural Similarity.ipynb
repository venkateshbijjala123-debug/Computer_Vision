```python
from skimage import io
from skimage.metrics import structural_similarity as ssim

# Load images (remembered path for original)
img_original = io.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')
img_modified = io.imread("image_ypbpr.jpg")

# Compute SSIM
ssim_original = ssim(img_original, img_original,
                     data_range=img_original.max() - img_original.min(),
                     channel_axis=-1)

ssim_different = ssim(img_original, img_modified,
                      data_range=img_modified.max() - img_modified.min(),
                      channel_axis=-1)

print(ssim_original, ssim_different)
```
