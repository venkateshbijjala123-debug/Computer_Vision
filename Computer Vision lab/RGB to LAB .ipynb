```python
from skimage import io, color
import matplotlib.pyplot as plt
import numpy as np

# Read image
img = io.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')

# Convert to Lab
img_lab = color.rgb2lab(img)

# Convert back to RGB
img_rgb = color.lab2rgb(img_lab)

# Normalize Lab for visualization (scale values into [0,1])
img_lab_norm = (img_lab - img_lab.min()) / (img_lab.max() - img_lab.min())

# Show both figures side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(img_lab_norm)
axes[0].set_title("Lab Image (normalized)")
axes[0].axis("off")

axes[1].imshow(img_rgb)
axes[1].set_title("RGB Image (from Lab)")
axes[1].axis("off")

plt.show()
```
