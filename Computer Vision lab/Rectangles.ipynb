```python
from skimage import io, draw
import matplotlib.pyplot as plt
import numpy as np

# Read image (remembered path)
img = io.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')

# Rectangle function
def rectangle(x, y, w, h):
    rr, cc = draw.polygon([x, x+w, x+w, x], [y, y, y+h, y+h])
    return rr, cc

# Draw rectangle safely within image bounds
rr, cc = rectangle(10, 10, min(200, img.shape[0]-20), min(200, img.shape[1]-20))
img[rr, cc] = 0  # black fill

# Show image
plt.imshow(img)
plt.axis("off")
plt.show()
```
