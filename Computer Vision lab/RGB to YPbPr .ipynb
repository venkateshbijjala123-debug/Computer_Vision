```python
# Import libraries
from skimage import io, color
import matplotlib.pyplot as plt
import numpy as np

# Read image (remembered path)
img = io.imread(r'C:\Users\venka\Downloads\Computer Vision lab\images.jpg')

# Convert to YPbPr
img_ypbpr = color.rgb2ypbpr(img)

# Convert back to RGB
img_rgb = color.ypbpr2rgb(img_ypbpr)

# Clip values to [0,1] to avoid warnings
img_ypbpr_clipped = np.clip(img_ypbpr, 0, 1)
img_rgb_clipped = np.clip(img_rgb, 0, 1)

# Show both figures side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(img_ypbpr_clipped)
axes[0].set_title("YPbPr Image (clipped)")
axes[0].axis("off")

axes[1].imshow(img_rgb_clipped)
axes[1].set_title("RGB Image (from YPbPr)")
axes[1].axis("off")

plt.show()
```
