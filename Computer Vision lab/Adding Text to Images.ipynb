```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(r'C:\Users\venka\Downloads\Computer Vision lab\IMG_1626.jpg')

# Define font
font = cv2.FONT_HERSHEY_SIMPLEX

# Add large text (black color)
cv2.putText(image, "Hello World!", (50, 250), font, 4.0, (0, 0, 0), 8, cv2.LINE_AA)

# Show image with matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
